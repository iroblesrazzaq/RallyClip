```python
from manual_court import CourtDetector
import cv2
import numpy as np
```

```python
# Initialize the detector
detector = CourtDetector()

# Process a video
playable_area_mask, processed_frame, metadata = detector.process_video("./raw_videos/Aditi Narayan ｜ Matchplay.mp4")
```

```python
# Create a colored mask overlay
# Convert the mask to 3-channel if it's grayscale
if len(playable_area_mask.shape) == 2:
    mask_colored = cv2.cvtColor(playable_area_mask, cv2.COLOR_GRAY2BGR)
else:
    mask_colored = playable_area_mask.copy()

# Create a green overlay for the playable area (you can change the color)
overlay_color = np.array([0, 255, 0], dtype=np.uint8)  # Green color
mask_overlay = np.zeros_like(processed_frame)
mask_overlay[playable_area_mask > 0] = overlay_color

# Blend the mask overlay with the processed frame
alpha = 0.3  # Transparency factor (0.0 = fully transparent, 1.0 = fully opaque)
masked_frame = cv2.addWeighted(processed_frame, 1 - alpha, mask_overlay, alpha, 0)

# Display the result
cv2.imshow('Processed Frame with Mask Overlay', masked_frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
